```python
import pandas as pd
import matplotlib.pyplot as plt

color = ['blue','green','red','pink']

plotdata = pd.DataFrame({

    "Precision":[0.01,0.014,0.014,0.01],

    "Recall":[0.954,0.898,0.855,0.954],

    "MAP":[0.415,0.358,0.321,0.257],
    
    "MAP - nội suy":[0.442,0.389,0.328,0.284]},

    index=["VSM", "LM-Uni", "LM_Bi", "Whoosh"])

ax = plotdata.plot(kind="bar",figsize=(15, 8),color=color,rot=0,width = 0.8,fontsize = 14)

plt.title("Compare VSM, LM and Whoosh")

plt.xlabel("Algorithms")

plt.ylabel("Score")

for container in ax.containers:
  for c in container:
    ax.annotate('{:.2f}'.format(c.get_height()),xy = (c.get_xy()[0] + 0.045,c.get_height() + 0.02),va = 'center')
```
